```python
#import pandas as pd
mine = pd.read_csv("mine.csv")
mine
```

```python
#head method is used to view the top records in a dataframe default is five but can take different parameters
mine.head()
```

```python
#tail view a given number of records from the bottom
mine.tail(3)
```

```python
#info checks information about the dataframe
mine.info()
```

```python
#Dropna() without any arguments drops any rows containing at least one missing value and returns a new dataframe
mine.dropna()
```

```python
#dropna(axis=1) drops columns instead of rows
mine.dropna(axis=1)
```

```python
#fillna() is used to fill missing values
#can take different parameters
# with specific value fillna(value)
# with mean of each column fillna(mine.mean)
# with forward fill method fillna(method ="fill")
mine.fillna(mine.mean)
mine
```

```python
work = pd.read_csv('Work.csv')
work
```

```python
work.info
```

```python
#Fill the missing data withthe mean value
work.fillna(work.mean)
work
```

```python
#Drop rows with missing data
work.dropna()
work
```

```python
#convert age column to numeric
work["age"] = pd.to_numeric(work["age"], errors = 'coerce')
work
```

```python
#Change py-score colum to numeric
work["py-score"] = pd.to_numeric(work["py-score"], errors = 'coerce')
work
```

```python
#drop duplicates based on 'ID' column
work = work.drop_duplicates()
work
```
